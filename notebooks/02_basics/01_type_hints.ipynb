````python
from typing import Optional

def find_user(user_id: int) -> Optional[dict]:
    """å°‹æ‰¾ç”¨æˆ¶ï¼Œæ‰¾ä¸åˆ°å›å‚³ None"""
    users = {1: {"name": "å°æ˜"}, 2: {"name": "å°è¯"}}
    return users.get(user_id)  # å¯èƒ½å›å‚³ dict æˆ– None

# Python 3.10+ å¯ä»¥ç”¨ | èªæ³•
def find_user_v2(user_id: int) -> dict | None:
    """åŒä¸Šï¼Œä½¿ç”¨æ–°èªæ³•"""
    users = {1: {"name": "å°æ˜"}, 2: {"name": "å°è¯"}}
    return users.get(user_id)

print(f"æ‰¾åˆ°ç”¨æˆ¶ 1: {find_user(1)}")
print(f"æ‰¾åˆ°ç”¨æˆ¶ 99: {find_user(99)}")
````

### Union - è¯åˆé¡å‹

`Union[T1, T2]` è¡¨ç¤ºå€¼å¯èƒ½æ˜¯ `T1` æˆ– `T2`ã€‚

````python
from typing import Union

def process_id(id_value: Union[int, str]) -> str:
    """è™•ç† IDï¼Œå¯ä»¥æ˜¯æ•´æ•¸æˆ–å­—ä¸²"""
    return f"ID-{id_value}"

# Python 3.10+ å¯ä»¥ç”¨ | èªæ³•
def process_id_v2(id_value: int | str) -> str:
    """åŒä¸Šï¼Œä½¿ç”¨æ–°èªæ³•"""
    return f"ID-{id_value}"

print(process_id(123))
print(process_id("ABC"))
````

### Callable - å¯å‘¼å«é¡å‹

````python
from typing import Callable

def apply_operation(
    x: int, 
    y: int, 
    operation: Callable[[int, int], int]
) -> int:
    """å°å…©å€‹æ•¸å­—å¥—ç”¨æŒ‡å®šçš„é‹ç®—"""
    return operation(x, y)

# å®šç¾©é‹ç®—å‡½å¼
def add(a: int, b: int) -> int:
    return a + b

def multiply(a: int, b: int) -> int:
    return a * b

print(f"3 + 5 = {apply_operation(3, 5, add)}")
print(f"3 * 5 = {apply_operation(3, 5, multiply)}")
````
