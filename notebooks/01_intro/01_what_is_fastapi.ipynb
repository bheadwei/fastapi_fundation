### 1️⃣ Starlette - 高效能 ASGI 框架

#### 🤔 什麼是 ASGI？

**ASGI**（Asynchronous Server Gateway Interface）是 Python 的**非同步伺服器介面標準**。

| 比較 | WSGI（舊標準） | ASGI（新標準） |
|------|---------------|---------------|
| 處理方式 | 同步（一次處理一個請求） | 非同步（可同時處理多個請求） |
| WebSocket | ❌ 不支援 | ✅ 支援 |
| 效能 | 🐢 較慢 | ⚡ 更快 |
| 代表框架 | Flask, Django (舊版) | FastAPI, Django 4.0+ |

> 💡 **簡單理解**：想像餐廳服務生，WSGI 是一次只能服務一桌客人，ASGI 是可以同時照顧多桌客人！

#### Starlette 提供的功能

| 功能 | 說明 | 實際用途 |
|------|------|----------|
| 🌐 **Web 路由** | URL 對應到處理函式 | `/users` → `get_users()` |
| 🔌 **WebSocket** | 即時雙向通訊 | 聊天室、即時通知 |
| 📡 **背景任務** | 非同步執行耗時工作 | 發送 Email、處理檔案 |
| 🍪 **Session & Cookie** | 用戶狀態管理 | 登入狀態、購物車 |

---

#### 🍪 什麼是 Cookie 和 Session？

**Cookie（餅乾）** 🍪
- 儲存在**瀏覽器**中的小型資料
- 每次請求時會自動帶給伺服器
- 例如：記住登入狀態、網站偏好設定

**Session（會話）** 📝
- 儲存在**伺服器**中的用戶資料
- 透過 Cookie 中的 Session ID 來識別用戶
- 例如：購物車內容、登入用戶資訊

### 2️⃣ Pydantic - 資料驗證庫

#### Pydantic 提供的功能

| 功能 | 說明 | 實際用途 |
|------|------|----------|
| ✅ **資料驗證** | 自動檢查資料格式是否正確 | 確保 email 格式正確 |
| 📝 **資料序列化** | 物件 ↔ JSON 相互轉換 | API 回應格式化 |
| 🔧 **設定管理** | 讀取環境變數和設定檔 | 資料庫連線設定 |
| 📋 **JSON Schema** | 自動生成資料結構描述 | API 文檔自動生成 |

---

#### 📝 什麼是資料序列化？

**序列化（Serialization）** 就是把程式中的物件轉換成可以儲存或傳輸的格式（如 JSON）。

```python
# Python 物件（只有程式看得懂）
user = User(name="小明", age=18, email="ming@example.com")

# 序列化後的 JSON（人和程式都看得懂，可以傳輸）
{"name": "小明", "age": 18, "email": "ming@example.com"}
```

| 術語 | 方向 | 說明 |
|------|------|------|
| **序列化** | 物件 → JSON | 把 Python 物件轉成 JSON 字串 |
| **反序列化** | JSON → 物件 | 把 JSON 字串轉成 Python 物件 |

> 💡 **簡單理解**：序列化就像把立體的樂高模型拍成平面照片方便傳送，反序列化就是看著照片重新組裝樂高！

---

#### ✅ 什麼是資料驗證？

**資料驗證** 就是檢查收到的資料是否符合預期格式。

```python
# 定義資料格式
class User(BaseModel):
    name: str          # 必須是字串
    age: int           # 必須是整數
    email: EmailStr    # 必須是有效的 Email 格式

# ✅ 正確的資料 → 通過驗證
{"name": "小明", "age": 18, "email": "ming@example.com"}

# ❌ 錯誤的資料 → 驗證失敗，自動回傳錯誤訊息
{"name": "小明", "age": "十八歲", "email": "這不是email"}
```

> 💡 **為什麼重要？** 沒有驗證的話，錯誤資料可能導致程式崩潰或安全漏洞！
