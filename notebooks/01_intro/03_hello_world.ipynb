### CRUD å°æ‡‰

**CRUD** æ˜¯è³‡æ–™æ“ä½œçš„å››å¤§åŸºæœ¬å‹•ä½œï¼š

| CRUD | å…¨å | HTTP æ–¹æ³• | èªªæ˜ |
|------|------|-----------|------|
| **C** | Create | POST | å»ºç«‹ |
| **R** | Read | GET | è®€å– |
| **U** | Update | PUT / PATCH | æ›´æ–° |
| **D** | Delete | DELETE | åˆªé™¤ |

````python
# HTTP æ–¹æ³•å¯¦éš›ç¯„ä¾‹
from fastapi import FastAPI
from fastapi.testclient import TestClient
from pydantic import BaseModel

app = FastAPI()

# æ¨¡æ“¬è³‡æ–™åº«
books_db = {
    1: {"id": 1, "title": "Python å…¥é–€", "price": 500},
    2: {"id": 2, "title": "FastAPI æ•™å­¸", "price": 600},
}

class Book(BaseModel):
    title: str
    price: int

# GET - è®€å–
@app.get("/books")
def get_all_books():
    """å–å¾—æ‰€æœ‰æ›¸ç±"""
    return list(books_db.values())

@app.get("/books/{book_id}")
def get_book(book_id: int):
    """å–å¾—å–®ä¸€æ›¸ç±"""
    return books_db.get(book_id, {"error": "æ‰¾ä¸åˆ°"})

# POST - å»ºç«‹
@app.post("/books")
def create_book(book: Book):
    """æ–°å¢æ›¸ç±"""
    new_id = max(books_db.keys()) + 1
    books_db[new_id] = {"id": new_id, **book.model_dump()}
    return books_db[new_id]

# PUT - å®Œæ•´æ›´æ–°
@app.put("/books/{book_id}")
def update_book(book_id: int, book: Book):
    """å®Œæ•´æ›´æ–°æ›¸ç± (éœ€è¦æ‰€æœ‰æ¬„ä½)"""
    books_db[book_id] = {"id": book_id, **book.model_dump()}
    return books_db[book_id]

# DELETE - åˆªé™¤
@app.delete("/books/{book_id}")
def delete_book(book_id: int):
    """åˆªé™¤æ›¸ç±"""
    if book_id in books_db:
        del books_db[book_id]
        return {"message": f"æ›¸ç± {book_id} å·²åˆªé™¤"}
    return {"error": "æ‰¾ä¸åˆ°"}

# æ¸¬è©¦
client = TestClient(app)

print("ğŸ“š HTTP æ–¹æ³•ç¤ºç¯„\n")

print("GET /books - å–å¾—æ‰€æœ‰æ›¸ç±")
print(client.get("/books").json())
print()

print("GET /books/1 - å–å¾— ID=1 çš„æ›¸")
print(client.get("/books/1").json())
print()

print("POST /books - æ–°å¢æ›¸ç±")
print(client.post("/books", json={"title": "Django æŒ‡å—", "price": 550}).json())
print()

print("PUT /books/1 - å®Œæ•´æ›´æ–°æ›¸ç±")
print(client.put("/books/1", json={"title": "Python é€²éš", "price": 700}).json())
print()

print("DELETE /books/2 - åˆªé™¤æ›¸ç±")
print(client.delete("/books/2").json())
````
